```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_scaled= StandardScaler().fit_transform(df_market_data[["price_change_percentage_24h", "price_change_percentage_7d",
                                                 "price_change_percentage_14d",	"price_change_percentage_30d",
                                                 "price_change_percentage_60d", "price_change_percentage_200d",
                                                 "price_change_percentage_1y"]]
)
```

```python
# Create a DataFrame with the scaled data
market_data_df=pd.DataFrame(df_market_scaled, columns= ["price_change_percentage_24h", "price_change_percentage_7d",
                                                 "price_change_percentage_14d",	"price_change_percentage_30d",
                                                 "price_change_percentage_60d", "price_change_percentage_200d",
                                                 "price_change_percentage_1y"])

# Copy the crypto names from the original DataFrame
market_data_df["coin_id"] = df_market_data.index

# Set the coin_id column as index
market_data_df = market_data_df.set_index("coin_id")


# Display the scaled DataFrame
market_data_df.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_value=list(range(1,11))
```

```python
# Create an empty list to store the inertia values

inertia=[]
# Create a for loop to compute the inertia with each possible value of k
for i in k_value:
    
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(market_data_df)
    inertia.append(k_model.inertia_)
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_value, "inertia": inertia}


# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
df_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xticks=k_value
)
# the different values of k to visually identify the optimal value for k.
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** After using the elbow curve we can see that best value for k is 4, this is the point where the elbow is created. Ideally 4 clusters will give us a good balance with the data without adding too many data points. 

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(market_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_predictions = model.predict(market_data_df)

# Print the resulting array of cluster values.
k_predictions
```

```python
# Create a copy of the scaled DataFrame
market_predictions_df=market_data_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
market_predictions_df['Cluster'] = k_predictions

# Display the copy of the scaled DataFrame
market_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting

# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.

# Color the graph points with the labels found using K-Means and

# add the crypto name in the `hover_cols` parameter to identify

# the cryptocurrency represented by each data point.
market_predictions_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c='Cluster',  # Color by cluster labels
    cmap='Category10',
    legend="top",
    hover_cols=['coin_id'],  # Add coin_id to hover info
    title='Cryptocurrency Clusters'
).opts(yformatter="%.0f")
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca=PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
crypto_pca = pca.fit_transform(market_data_df)
# down to three principal components.

crypto_pca[:3]
# View the scaled PCA data
```

```python
# Retrieve the explained variance to determine how much information

# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is typically a value between 0 and 1 but expressed in a percentage, which indicates the proportion of the total variance in the data that is explained by those components. We add up to the components to get the total variance (Total Explained Variance=PC1+PC2+PC3). In our example above the tottal variance of the three principal components is approximately 0.8950 or 89.50%. 

```python
print(market_data_df.columns)
```

```python
# Create a new DataFrame with the PCA data.

pca_df = pd.DataFrame(crypto_pca, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original scaled DataFrame

pca_df["coin_id"] = df_market_data.index
# Set the coin_id column as index

pca_df= pca_df.set_index('coin_id')
# Display the scaled PCA DataFrame
pca_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values=list(range(1,11))
```

```python
# Create an empty list to store the inertia values

inertia=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k_values:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve

elbow_curve = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow = pd.DataFrame(elbow_curve)

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_elbow.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve",
    xticks=k_values
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The optimal k_value is 4. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** It does not differ from the original differ it remains consistent whether using the PCA-transformed data or the original data, indicating that the clustering structure is stable across both datasets.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model_4 = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model_4.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
k_pca_predictions = model_4.predict(pca_df)

# Print the resulting array of cluster values.
k_pca_predictions


```

```python
# Create a copy of the scaled PCA DataFrame

scaled_pca_df=pca_df.copy()
# Add a new column to the copy of the PCA DataFrame with the predicted clusters

scaled_pca_df["PCA Predictions"]=k_pca_predictions

# Display the copy of the scaled PCA DataFrame
scaled_pca_df.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.

scaled_pca_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by = 'PCA Predictions',
    hover_cols='coin_id'
)
```

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** The benefit seems clear in that it simplyfies the data so that you are looking a basically a summary or synthesis of the data by calling on unrelated variables and grouping them together by variance. These principal components are forced into a linear relationship in which otherwise they might not be based on raw data. If you are dealing with large data sets and you want to make quick assumptions, this is a useful tool. It captures the most variance in the data, but at the same time it sacrificies accuracy and it might even make it harder to understand the real world impact of what you are looking at.
